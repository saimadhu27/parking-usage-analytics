```python
def fill_missing_times(df):

    # Fill EntranceTime where missing and ExitTime exists
    mask_entrance_missing = df['EntranceTime'].isna() & df['ExitTime'].notna()
    overnight_mask = df['Overnight'] == 1

    # Filling entry time when overnight = 1
    df.loc[mask_entrance_missing & overnight_mask, 'EntranceTime'] = (
        df.loc[mask_entrance_missing & overnight_mask, 'ExitTime'].dt.normalize() - pd.Timedelta(days=1)
    )
    
    # Filling entry time when overnight != 1
    df.loc[mask_entrance_missing & ~overnight_mask, 'EntranceTime'] = (
        df.loc[mask_entrance_missing & ~overnight_mask, 'ExitTime'].dt.normalize()
    )

    # Fill ExitTime where missing and EntranceTime exists
    mask_exit_missing = df['ExitTime'].isna() & df['EntranceTime'].notna()

    # Filling exit time when overnight = 1
    df.loc[mask_exit_missing & overnight_mask, 'ExitTime'] = (
        df.loc[mask_exit_missing & overnight_mask, 'EntranceTime'].dt.normalize() + pd.Timedelta(days=1)
    )
    
    # Filling exit time when overnight = 1
    df.loc[mask_exit_missing & ~overnight_mask, 'ExitTime'] = (
        df.loc[mask_exit_missing & ~overnight_mask, 'EntranceTime'].dt.normalize() + pd.Timedelta(hours=23, minutes=59, seconds=59)
    )

    return df

```

```python
card_df_c = fill_missing_times(card_df)
card_df_c
```

```python
# Calculate duration in hours as float
card_df_c['DurationHours'] = (card_df_c['ExitTime'] - card_df_c['EntranceTime']) / pd.Timedelta(hours=1)
card_df_c['DurationHours'] = card_df_c['DurationHours'].round(2)
```

```python
#card_df_c.to_csv('cardtransaction_cleaned.csv', index=False)
```

```python
df = card_df_c.copy()
```

```python
lot_numbers = df['LotNumber'].unique()
results = []

for lot in tqdm(lot_numbers, desc='Processing lots'):
    lot_data = df[df['LotNumber'] == lot][['EntranceTime', 'ExitTime']].copy()
    
    # Build events: +1 on entry, -1 on exit
    entries = pd.DataFrame({
        'Time': lot_data['EntranceTime'],
        'Change': 1
    })
    exits = pd.DataFrame({
        'Time': lot_data['ExitTime'],
        'Change': -1
    })
    
    events = pd.concat([entries, exits]).sort_values('Time')
    events['SimultaneousUsage'] = events['Change'].cumsum()
    
    # Optional: round to hour if you want hourly grid
    events['TimePoint'] = events['Time'].dt.floor('H')
    
    # Keep max per hour
    hourly_usage = events.groupby('TimePoint')['SimultaneousUsage'].max().reset_index()
    hourly_usage['LotNumber'] = lot
    
    results.append(hourly_usage[['LotNumber', 'TimePoint', 'SimultaneousUsage']])

# Combine all lots
simultaneous_df = pd.concat(results, ignore_index=True)
```
